```python
df_copy = df.copy(deep=True)
```

```python
df.drop(['ID','Year Type','Employee Identifier','Job Code','Job','Organization Group Code'],axis=1,inplace=True)
```

```python
cat_data = ['Organization Group','Department Code','Department','Union','Job Family Code','Job Family']
for col in cat_data:
    df[col] = df[col].astype('category') 
```

```python
df.info()
```

```python
neg_col=['Overtime','Other Salaries','Other Salaries','Retirement','Health/Dental','Other Benefits','Total Benefits']
for i in neg_col:
    print(df[df[i] < 0].shape)
```

```python
for j in neg_col:
    ind=df[(df[j]<0) & (df[j] == 1)].index
    df.drop(ind,axis=0,inplace=True)
```

```python
neg_col=['Overtime','Other Salaries','Other Salaries','Retirement','Health/Dental','Other Benefits','Total Benefits']
for i in neg_col:
    print(df[(df[j]<0) & (df[j] == 1)].shape)
```

```python
def replace(col):
    median, std = col.median(), col.std()  #Get the median and the standard deviation of every group 
    outliers = (col - median).abs() > 2*std # Subtract median from every member of each group. Take absolute values > 2std
    col[outliers] = col.median()       
    return col
```

```python
float_cols = ['Overtime','Other Salaries','Other Salaries','Retirement','Health/Dental','Other Benefits','Total Benefits']
for i in float_cols:
    df[i].transform(replace)
```

```python
DF_test = df[df['train_test']==0]
```

```python
DF_train = df[df['train_test']==1]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
DF_test
```

```python
DF_test.drop('Class',axis=1,inplace=True)
```

```python
X1 = DF_train.drop('Class',axis=1)
y1 = DF_train['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X1, y1, random_state = 1,test_size=0.2)
```

```python
X_train.shape,DF_test.shape
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
DF_test = sc.transform(DF_test)
```

```python
X_df=DF_train.drop('Class',axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model=DecisionTreeClassifier(criterion='entropy',random_state=0) 
dt_model.fit(X_train,y_train)
```

```python
imp_df = pd.DataFrame(dt_model.feature_importances_, columns = ["Imp"], index = X_df.columns).sort_values('Imp',ascending=False)
imp_df.head(20)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=150)
pca.fit(X_train)
X_pca = pca.transform(X_train)
   
pca.components_
#pca.explained_variance_
pca.explained_variance_ratio_
pca.explained_variance_ratio_.cumsum()   
```

```python
from sklearn.ensemble import BaggingClassifier,GradientBoostingClassifier,RandomForestClassifier,ExtraTreesClassifier
from xgboost import XGBClassifier
from sklearn import metrics
from sklearn.metrics import classification_report,accuracy_score
```

```python
model1 = XGBClassifier()
model2 = ExtraTreesClassifier()
model3 = RandomForestClassifier()
model4 = GradientBoostingClassifier()
model5 = BaggingClassifier()
model1.fit(X_pca,y_train)
model2.fit(X_pca,y_train)
model3.fit(X_pca,y_train)
model4.fit(X_pca,y_train)
model5.fit(X_pca,y_train)
```

```python
Xtest_pca = pca.transform(X_test)
y_pred1 = model1.predict(Xtest_pca)
print('XGBClassifier Accuracy = ',accuracy_score(y_test,y_pred1))
```

```python
Xtest_pca = pca.transform(X_test)
y_pred2 = model2.predict(Xtest_pca)
print('ExtraTreesClassifier Accuracy = ',accuracy_score(y_test,y_pred2))
```

```python
Xtest_pca = pca.transform(X_test)
y_pred3 = model3.predict(Xtest_pca)
print('RandomForestClassifier Accuracy = ',accuracy_score(y_test,y_pred3))
```

```python
Xtest_pca = pca.transform(X_test)
y_pred4 = model4.predict(Xtest_pca)
print('GradientBoostingClassifier Accuracy = ',accuracy_score(y_test,y_pred4))
```

```python
Xtest_pca = pca.transform(X_test)
y_pred5 = model5.predict(Xtest_pca)
print('BaggingClassifier Accuracy = ',accuracy_score(y_test,y_pred5))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred1))
print(classification_report(y_test, y_pred2))
print(classification_report(y_test, y_pred3))
print(classification_report(y_test, y_pred4))
print(classification_report(y_test, y_pred5))
```
